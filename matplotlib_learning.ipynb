```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# resize graph
# plt.figure(figsize=(3,2),dpi=200)

# plt.plot(x,y,label='2x',color='y',linewidth=2,linestyle='--',marker='.',markersize=10,markeredgecolor='green')
# shorthand of above code
# fnt=[color][marker][line]
plt.plot(x,y,'y^--',label='2x')

# line 2
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:7],x2[:7]**2,'g^--',label='x^2')
plt.plot(x2[6:],x2[6:]**2,'g^-',label='x^2')


plt.title("First Graphh",fontdict={'fontname':"Comic Sans MS",'fontsize':20})
plt.xlabel("xxxxxxx")
plt.ylabel("yyyyyyy")

plt.xticks([0,1,2,3,4,5])#custom x and y ticks
# plt.yticks([0,2,4,6,8,10])

plt.legend()#to show 2x label

plt.savefig('mygraph.png')



plt.show()
```

```python
# barchart
lables=['A','B','C']
vals=[3,6,2]

bars = plt.bar(lables,vals)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('.')

plt.figure(figsize=(6,4))

plt.show()
```
